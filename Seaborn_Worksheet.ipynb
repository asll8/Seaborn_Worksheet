```
sns.kdeplot(x, shade = True)
```

##FacetGrid

```
sns.FacetGrid(row = "smoker", col = "time", data= tips)
```

```
x = sns.FacetGrid(row="smoker", col="time", data=tips)
x.map(plt.hist, "total_bill")
```

```
x = sns.FacetGrid(row="smoker", col="time", data=tips)
x.map(plt.scatter, "total_bill", "tip")
```

```
x = sns.FacetGrid(row="smoker", col="time", data=tips)
x.map(sns.regplot, "total_bill", "tip")
```

```
x = sns.FacetGrid(hue="smoker", col="time", data=tips)
x.map(plt.scatter, "total_bill", "tip")
x.add_legend()
```

##Regression Plot

```
sns.regplot(x = "tip", y = "total_bill", data = tips)
```

```
mean = [0, 0]
cov = [[1, 0.9], [0.9, 1]]
x, y = np.random.multivariate_normal(mean, cov, 500).T
```

```
sns.regplot(x,y)
```

```
mean, cov = [2, 10], [(1, .9), (.9, 1)]
x, y = np.random.multivariate_normal(mean, cov, size = 400).T
sns.regplot(x, y)
```

```
mean, cov = [2, 10], [(1, .9), (.9, 1)]
x, y = np.random.multivariate_normal(mean, cov, size = 400).T
mean2, cov2 = [5, 18], [(1, .2), (.2, 1)]
x2, y2 = np.random.multivariate_normal(mean2, cov2, size = 400).T
sns.regplot(x, y)
sns.regplot(x2, y2)
```
